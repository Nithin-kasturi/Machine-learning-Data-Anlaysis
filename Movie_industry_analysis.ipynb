```python
df_normalized=df
for cols in df_normalized.columns:
    if(df_normalized[cols].dtype=='object'):
        df_normalized[cols]=df_normalized[cols].astype('category')
        df_normalized[cols]=df_normalized[cols].cat.codes
        
```

```python
df_normalized
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(df_normalized.corr(method="pearson"),annot=True)
# plt.figure(figsize=(32,32))
```

```python
correlation_matrix=df_normalized.corr()
correlation_pairs=correlation_matrix.unstack()
correlation_pairs
```

```python
sorted_pairs=correlation_pairs.sort_values()
sorted_pairs
```

```python
high_correlation=sorted_pairs[(sorted_pairs)>0.5]
```
