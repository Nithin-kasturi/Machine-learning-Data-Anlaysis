```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('tracks.csv')
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
sorted_df=df.sort_values('popularity',ascending=False)
```

```python
sorted_df.head(10)
```

```python
most_popular=df.query('popularity>90',inplace=False).sort_values('popularity',ascending=True)
```

```python
most_popular
```

```python
df.set_index('release_date',inplace=True)
```

```python
df.index=pd.to_datetime(df.index)
```

```python
df
```

```python
df.rename(columns={'duration_ms':'Duration'},inplace=True)
```

```python
df.info()
```

```python
df['Duration']=df['Duration'].apply(lambda x:round(x/1000))
```

```python
df["Duration"]
```

```python
df.info()
```

```python
correlation=df.drop(['explicit','key','mode'],axis=1).corr(method='pearson')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(correlation,annot=True,vmin=-1,vmax=1)
```

```python
sample_df=df.sample(int(0.004*len(df)))
```

```python
print(len(sample_df))
```

```python
sns.regplot(data=sample_df,x='energy',y='loudness',color='c')
plt.figure(figsize=(14,6))
```

```python
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='b')
```
